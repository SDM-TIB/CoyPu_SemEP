### Execute the query and store data into a pandas Dataframe

```python
world_bank = []

sparql.setQuery(query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()


for r in results['results']['bindings']:
    row = {'country': r['country']['value'].replace(',', ''),
           'indicator': r['label_indicator']['value'].replace(',', ''),
           'value': r['value']['value']}
    
    world_bank.append(row)

world_bank = pd.DataFrame.from_dict(world_bank)
world_bank = world_bank.drop_duplicates()
replacement_mapping_dict = dict(zip(df.s, df.o))
world_bank['country'].replace(replacement_mapping_dict, inplace=True)
world_bank = world_bank[world_bank.country.str.contains("https://data.coypu.org/country/") == True]
display(world_bank.shape, world_bank.head(2))
```

```python
df_1 = world_bank[world_bank.country.str.contains("https://data.coypu.org/country/") == False]
# df_1.country.unique()
```

```python
missing_link = []
for s in df_1.country.unique():    
    query = """SELECT DISTINCT ?label_country WHERE {
        <"""+s+"""> <http://www.w3.org/2000/01/rdf-schema#label> ?label_country .
    }
    """
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    for r in results['results']['bindings']:
        row = {'country': s, 'label_country': r['label_country']['value']}
        missing_link.append(row)

missing_link = pd.DataFrame.from_dict(missing_link)
missing_link
```

### categories: 'low', 'mediumLow', 'medium', 'mediumHigh', 'high'. (Higher is better)

```python
country_ind = pd.DataFrame()
for indicator in world_bank.indicator.unique().tolist():
    lower = False
    if indicator in ['CO2 emissions (metric tons per capita)', 'Incidence of tuberculosis (per 100000 people)',
                    'Prevalence of anemia among pregnant women (%)',
                     'Incidence of HIV all (per 1000 uninfected population)',
                     'Prevalence of current tobacco use (% of adults)']:
        lower = True
    country_ind = pd.concat([country_ind, Utility.convert_to_category(world_bank, indicator, lower)])
country_ind = country_ind[['country', 'category_indicator', 'indicator']]
country_ind
```

## From icews_endpoint

```python
sparql = SPARQLWrapper('https://labs.tib.eu/sdm/icews_endpoint/sparql')
```

```python
query = """
PREFIX coy: <https://schema.coypu.org/global#>
CONSTRUCT {?s a coy:SocioPoliticalEvent .
  ?s coy:hasCountryLocation ?hasCountryLocation .
  ?s coy:hasTargetCountry ?hasTargetCountry .
  ?s coy:hasDate ?hasDate .
  ?s coy:hasTargetSectors ?hasTargetSectors .
  ?s coy:hasTarget ?hasTarget .
  ?s coy:hasIntensity ?hasIntensity .
  ?s coy:hasSourceSectors ?hasSourceSectors .
  ?s coy:hasSourceCountry ?hasSourceCountry}
WHERE {
  ?s a coy:SocioPoliticalEvent .
  ?s coy:hasCountryLocation ?hasCountryLocation .
  #optional{?s coy:hasStoryId ?hasStoryId .}
  optional{?s coy:hasTargetCountry ?hasTargetCountry .}
  optional{?s coy:hasDate ?hasDate .}
  optional{?s coy:hasTargetSectors ?hasTargetSectors .}
  optional{?s coy:hasTarget ?hasTarget .}
  optional{?s coy:hasIntensity ?hasIntensity .}
  optional{?s coy:hasSourceSectors ?hasSourceSectors .}
  optional{?s coy:hasSourceCountry ?hasSourceCountry}
}"""
icews = []
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

for r in results['results']['bindings']:
    row = {'s': r['s']['value'], 'p': r['p']['value'], 'o': r['o']['value']}
    icews.append(row)

icews = pd.DataFrame.from_dict(icews)
# icews = adding_prefix(icews)
icews = filter_by_year(icews, 2022)
display(icews.shape, icews.head(2))
```

### Get Country and sameAs

```python
q_CountryLocation = """
PREFIX coy: <https://schema.coypu.org/global#>
SELECT DISTINCT ?s ?o
WHERE {
  ?SPE a coy:SocioPoliticalEvent .
  ?SPE coy:hasCountryLocation ?s .
  ?s <http://www.w3.org/2002/07/owl#sameAs> ?o .
}"""
q_TargetCountry = """
PREFIX coy: <https://schema.coypu.org/global#>
SELECT DISTINCT ?s ?o
WHERE {
  ?SPE a coy:SocioPoliticalEvent .
  ?SPE coy:hasTargetCountry ?s .
  ?s <http://www.w3.org/2002/07/owl#sameAs> ?o  .
}"""
q_SourceCountry = """
PREFIX coy: <https://schema.coypu.org/global#>
SELECT DISTINCT ?s ?o
WHERE {
  ?SPE a coy:SocioPoliticalEvent .
  ?SPE coy:hasSourceCountry ?s .
  ?s <http://www.w3.org/2002/07/owl#sameAs> ?o
}"""
df = pd.concat([read_result_country(q_CountryLocation, sparql), read_result_country(q_TargetCountry, sparql),
                read_result_country(q_SourceCountry, sparql)])
df = df.drop_duplicates()

df_temp = pd.DataFrame(data=[['the%20former%20Yugoslav%20Republic%20of%20Macedonia','MKD'],['Egypt','EGY'],
                             ['Kyrgyzstan','KGZ'], ['Moldova%2C%20Republic%20of', 'MDA'], ['Saint%20Lucia', 'LCA'],
                             ['Slovakia', 'SVK'], ['Turkey', 'TUR'], ['Yemen', 'YEM']],columns=['s', 'o'])
df_temp['s'] = 'https://data.coypu.org/country/' + df_temp['s'].astype(str)
df_temp['o'] = 'https://data.coypu.org/country/' + df_temp['o'].astype(str)

df = pd.concat([df, df_temp])
display(df.shape, df)
```

### Replacement URI Country by Country Code
